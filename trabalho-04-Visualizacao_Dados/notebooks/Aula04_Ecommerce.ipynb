````python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Dataset: Olist Brazilian E-Commerce
# URL: https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce

# Define o caminho do arquivo que quer carregar (exemplo: pedidos)
file_path = "olist_orders_dataset.csv"

# Carrega direto em um DataFrame
orders = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    file_path
)

print("Pedidos:", orders.shape)
display(orders.head())
````

````python
customers = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    "olist_customers_dataset.csv"
)

reviews = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    "olist_order_reviews_dataset.csv"
)

items = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    "olist_order_items_dataset.csv"
)

products = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    "olist_products_dataset.csv"
)

sellers = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    "olist_sellers_dataset.csv"
)

orders = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "olistbr/brazilian-ecommerce",
    "olist_orders_dataset.csv"
)

````

````python

payments = None
if os.path.exists("olist_order_payments_dataset.csv"):
    payments = pd.read_csv("olist_order_payments_dataset.csv")

print(
    orders.shape, customers.shape, reviews.shape,
    items.shape, products.shape, sellers.shape,
    (payments.shape if payments is not None else "sem payments")
)
````

````python
# Converter datas do orders/reviews (coerÃ§Ã£o de invÃ¡lidos para NaT)
date_cols_orders = [
    "order_purchase_timestamp","order_approved_at",
    "order_delivered_carrier_date","order_delivered_customer_date",
    "order_estimated_delivery_date"
]
orders[date_cols_orders] = orders[date_cols_orders].apply(pd.to_datetime, errors="coerce")

date_cols_reviews = ["review_creation_date","review_answer_timestamp"]
for c in date_cols_reviews:
    if c in reviews.columns:
        reviews[c] = pd.to_datetime(reviews[c], errors="coerce")

# Duplicatas por chaves naturais
print("Dup orders (order_id):", orders.duplicated("order_id").sum())
print("Dup customers (customer_id):", customers.duplicated("customer_id").sum())
print("Dup reviews (review_id):", reviews.duplicated("review_id").sum() if "review_id" in reviews.columns else 0)
print("Dup items (order_id + order_item_id):", items.duplicated(["order_id","order_item_id"]).sum())
print("Dup products (product_id):", products.duplicated("product_id").sum())
print("Dup sellers (seller_id):", sellers.duplicated("seller_id").sum())
if payments is not None:
    print("Dup payments (order_id + payment_sequential):",
          payments.duplicated(["order_id","payment_sequential"]).sum())

# Remover duplicatas
orders    = orders.drop_duplicates("order_id").copy()
customers = customers.drop_duplicates("customer_id").copy()
if "review_id" in reviews.columns:
    reviews = reviews.drop_duplicates("review_id").copy()
items     = items.drop_duplicates(["order_id","order_item_id"]).copy()
products  = products.drop_duplicates("product_id").copy()
sellers   = sellers.drop_duplicates("seller_id").copy()
if payments is not None:
    payments = payments.drop_duplicates(["order_id","payment_sequential"]).copy()

# Nulos chave
print("\nNulos - orders:\n", orders[date_cols_orders + ["order_status"]].isnull().sum())
print("\nNulos - reviews:\n", reviews[["review_score"]].isnull().sum())
print("\nNulos - items:\n", items[["price","freight_value"]].isnull().sum())
````

````python
# (A) orders + items â†’ traz frete, preÃ§o, seller, product
df = items.merge(orders, on="order_id", how="left", suffixes=("_item","_order"))

# (B) + products â†’ categoria
df = df.merge(products[["product_id","product_category_name"]], on="product_id", how="left")

# (C) + sellers â†’ info do seller (cidade/estado do seller, Ãºtil p/ anÃ¡lises futuras)
df = df.merge(sellers, on="seller_id", how="left", suffixes=("","_seller"))

# (D) + customers â†’ info do cliente (unique_id p/ recompra, cidade/estado do cliente)
df = df.merge(customers[["customer_id","customer_unique_id","customer_city","customer_state"]],
              on="customer_id", how="left")

# (E) + reviews â†’ avaliaÃ§Ã£o do cliente no pedido
df = df.merge(reviews[["order_id","review_score","review_creation_date"]], on="order_id", how="left")

# (F) + payments (se existir) â†’ parcelas e tipo de pagamento (agregamos por pedido)
if payments is not None:
    pay_agg = payments.groupby("order_id").agg(
        payment_installments=("payment_installments","max"),   # maior nÂº de parcelas no pedido
        payment_types=("payment_type", lambda x: ",".join(sorted(set(x))))
    ).reset_index()
    df = df.merge(pay_agg, on="order_id", how="left")

print("Mestre (nÃ­vel item):", df.shape)
df.head(3)
````

````python
# Prazo de entrega (dias) â€” sÃ³ faz sentido para delivered
df["prazo_entrega"] = (
    df["order_delivered_customer_date"] - df["order_purchase_timestamp"]
).dt.days

# Atraso em dias (pode ser <0 se entregou antes da estimada)
df["atraso_dias"] = (
    df["order_delivered_customer_date"] - df["order_estimated_delivery_date"]
).dt.days

# Flag de atraso > 0
df["atraso_flag"] = (df["atraso_dias"] > 0).astype(int)

# Review positiva (4 ou 5)
df["positive_review"] = (df["review_score"] >= 4).astype(int)

# ReputaÃ§Ã£o do seller = mÃ©dia de review_score por seller (via pedidos avaliados)
seller_score = df.groupby("seller_id")["review_score"].mean()
df["seller_avg_review"] = df["seller_id"].map(seller_score)

# Recompra (nÃ­vel cliente): 1 se customer_unique_id tem 2+ pedidos diferentes
orders_per_customer = df.groupby("customer_unique_id")["order_id"].nunique()
recompra_map = (orders_per_customer > 1).astype(int)
df["recompra_flag"] = df["customer_unique_id"].map(recompra_map)

# Se payments estiver ausente, criamos coluna vazia para manter esquema consistente
if payments is None and "payment_installments" not in df.columns:
    df["payment_installments"] = np.nan

df.head(3)
````

````python
# Duplicatas jÃ¡ foram removidas nas bases de origem; aqui garantimos por seguranÃ§a:
df = df.drop_duplicates(subset=["order_id","order_item_id","product_id"])

# Nulos (decisÃµes)
df["product_category_name"] = df["product_category_name"].fillna("unknown")

# Outliers em prazo de entrega (somente delivered e com prazo vÃ¡lido)
mask_entregues = (df["order_status"] == "delivered") & df["prazo_entrega"].notna()
prazo = df.loc[mask_entregues, "prazo_entrega"]

# IQR (k = 1.0) â€” mais restritivo
Q1, Q3 = prazo.quantile(0.25), prazo.quantile(0.75)
IQR = Q3 - Q1
k = 1.0
low, high = Q1 - k*IQR, Q3 + k*IQR

# Cap suave (evitar remover linhas; mantemos para nÃ£o distorcer contagens por item)
df.loc[mask_entregues, "prazo_entrega"] = prazo.clip(lower=low, upper=high)

# Outliers em freight_value â€” cap por IQR no nÃ­vel item
frete = df["freight_value"].dropna()
Q1f, Q3f = frete.quantile(0.25), frete.quantile(0.75)
IQRf = Q3f - Q1f
lowf, highf = Q1f - 1.0*IQRf, Q3f + 1.0*IQRf
df["freight_value"] = df["freight_value"].clip(lower=lowf, upper=highf)
````

````python
# VersÃ£o 1 â€” NÃ­vel item (completo)
df_item = df.copy()

# VersÃ£o 2 â€” NÃ­vel pedido (agregar itens por pedido)
agg_dict = {
    "freight_value":"sum",
    "price":"sum",
    "payment_installments":"max",
    "prazo_entrega":"mean",
    "atraso_flag":"max",
    "review_score":"max",
    "positive_review":"max",
    "seller_avg_review":"mean"
}
cols_keep = [
    "order_id","customer_id","customer_unique_id","customer_city","customer_state",
    "order_status","order_purchase_timestamp","order_estimated_delivery_date",
    "order_delivered_customer_date"
]
df_order = df_item.groupby("order_id").agg(agg_dict).reset_index().merge(
    df_item[cols_keep].drop_duplicates("order_id"), on="order_id", how="left"
)

# VersÃ£o 3 â€” NÃ­vel cliente (uma linha por cliente Ãºnico)
df_customer = df_order.groupby("customer_unique_id").agg(
    n_pedidos=("order_id","nunique"),
    taxa_recompra=("order_id", lambda x: 1.0 if x.nunique()>1 else 0.0),
    media_review=("review_score","mean"),
    pct_reviews_positivas=("positive_review","mean"),
    prazo_medio=("prazo_entrega","mean"),
    atraso_pct=("atraso_flag","mean"),
    frete_medio=("freight_value","mean"),
).reset_index().merge(
    df_order[["customer_unique_id","customer_city","customer_state"]].drop_duplicates("customer_unique_id"),
    on="customer_unique_id", how="left"
)

# Salvar arquivos tratados
df_item.to_csv("olist_tratado_item.csv", index=False)
df_order.to_csv("olist_tratado_order.csv", index=False)
df_customer.to_csv("olist_tratado_customer.csv", index=False)

print("Salvos:",
      os.path.exists("olist_tratado_item.csv"),
      os.path.exists("olist_tratado_order.csv"),
      os.path.exists("olist_tratado_customer.csv"))
````

````python
# nÃ­vel item: review_score vem do pedido, mas categoria Ã© do item
cat_perf = (
    df_item[["product_category_name","review_score"]]
    .dropna()
    .groupby("product_category_name")
    .agg(n=("review_score","count"), media=("review_score","mean"),
         pct_pos=("review_score", lambda x: (x>=4).mean()))
    .query("n >= 200")
    .sort_values("pct_pos", ascending=False)
)

top10 = cat_perf.head(10)
bot10 = cat_perf.tail(10)

top10, bot10
````

Insight:
Categorias como mÃ³veis e telefonia tÃªm maiores Ã­ndices de insatisfaÃ§Ã£o, possivelmente por:

Problemas logÃ­sticos (produtos volumosos â†’ atrasos).

Expectativa vs realidade (mÃ³veis e decoraÃ§Ã£o).

Defeitos e suporte (telefonia e eletrÃ´nicos).


````python
# seller_avg_review jÃ¡ foi calculado no nÃ­vel item e propagado.
seller_perf = (
    df_item[["seller_id","seller_avg_review","review_score"]]
    .dropna()
    .groupby("seller_id")
    .agg(seller_avg=("seller_avg_review","mean"),
         n=("review_score","count"),
         pct_pos=("review_score", lambda x: (x>=4).mean()))
    .query("n >= 100")
)

# CorrelaÃ§Ã£o entre reputaÃ§Ã£o do seller e % de reviews positivas
seller_perf[["seller_avg","pct_pos"]].corr(method="spearman")
````

A reputaÃ§Ã£o do seller tem forte influÃªncia sobre a satisfaÃ§Ã£o dos clientes.
A correlaÃ§Ã£o de Spearman â‰ˆ 0,98 confirma que vendedores bem avaliados mantÃªm uma alta proporÃ§Ã£o de notas positivas (4â€“5).
Isso evidencia a importÃ¢ncia de polÃ­ticas de curadoria, acompanhamento e suporte aos sellers para garantir a qualidade do marketplace.
